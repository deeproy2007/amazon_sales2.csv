```python
import pandas as pd
df=pd.read_csv('unclean_sales_data_500.csv')
df.dropna(subset=['quantity'],inplace = True)
df['quantity']=pd.to_numeric(df['quantity'],errors='coerce')
a=df['quantity'].mean()
df.fillna({'quantity':a},inplace =True)
df['quantity']=df['quantity'].astype(int)
df['price'] =pd.to_numeric(df['price'],errors='coerce')
b=df['price'].mean()
df.fillna({'price':b},inplace =True)
df['price']=df['price'].astype(float)
df.dropna(subset=['category'],inplace =True)
df.dropna(subset=['state'],inplace=True)
df['state'].replace({'karnataka':'Karnataka','KA':'Karnataka','maharastra':'Maharastra','MH':'Maharastra'},inplace=True)
df['delivery_method'].replace({'cod':'COD','online':'Online','Onlne':'Online'},inplace=True)
df.dropna(subset=['delivery_method'],inplace =True)
df['order_date']=pd.to_datetime(df['order_date'],errors='coerce')
df.dropna(subset=['order_date'],inplace =True)
print(df.to_string())
df.to_csv('clean_sales_data_500.csv',index=False)
```
